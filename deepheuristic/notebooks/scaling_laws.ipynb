# Regularization

## Smoothness

```python
# For 35M model
labels = [0.001, 0.005, 0.01, 0.05, 0.1]
mae = [0.0353, 0.0352, 0.0350, 0.0392, 0.0515]
plt.plot(labels, mae, marker='o', label='Smoothness regularization')
plt.xlabel('Smoothness regularization')
plt.ylabel('MAE')
# Log2 scale for x-axis
plt.xscale('log')
plt.xticks(labels, labels)
plt.grid()
plt.title("Smoothness regularization vs MAE for 35M Model")
# Conclusion:Smoothness seems to help a lot for generalization
```

## Diversity

```python
# For 35M model
labels = [0.001, 0.005, 0.01, 0.05, 0.1]
mae = [0.0349, 0.0356, 0.0397, 0.0560, 0.0671]
plt.plot(labels, mae, marker='o', label='Diversity regularization')
plt.xlabel('Diversity regularization')
plt.ylabel('MAE')
# Log2 scale for x-axis
plt.xscale('log')
plt.xticks(labels, labels)
plt.grid()
plt.title("Diversity regularization vs MAE for 35M Model")
# Conclusion: There seems to have a clear trend the the less diversity regularization we have, the better the model is. However, the 0.001 value allows to have a better generalization than the baseline. So we can set it to 0.001 for the next experiments.
```

## Var

```python
# For 35M model
labels = [0.001, 0.005, 0.01, 0.05, 0.1]
mae = [0.0350, 0.0352, 0.0351, 0.0352, 0.0348]
plt.plot(labels, mae, marker='o', label='Variance regularization')
plt.xlabel('Variance regularization')
plt.ylabel('MAE')
# Log2 scale for x-axis
plt.xscale('log')
plt.xticks(labels, labels)
plt.grid()
plt.title("Variance regularization vs MAE for 35M Model")
# Conclusion: There is no clear trend. However, the 0.1 value clearly helps the generalization in test and in val. We can set it to 0.1 for the next experiments.
```

## Orthogonality

```python
# For 35M model
labels = [0.001, 0.005, 0.01, 0.05, 0.1]
mae = [0.0352, 0.0350, 0.0350, 0.0349, 0.0351]
plt.plot(labels, mae, marker='o', label='Orthogonality regularization')
plt.xlabel('Orthogonality regularization')
plt.ylabel('MAE')
# Log2 scale for x-axis
plt.xscale('log')
plt.xticks(labels, labels)
plt.grid()
plt.title("Orthogonality regularization vs MAE for 35M Model")
# Conclusion: There is a clear trends that the more orthogonality we have, the better the model generalizes up to 0.05.
```
