```python
import matplotlib.pyplot as plt
import numpy as np
```

# Projection Dim
Analysis of the size of the projection dimension on the performance of a model.

```python
# For the 8M model
labels = [64, 128, 256, 512]
mae = [0.0608, 0.0420, 0.0468, 0.0401]
plt.plot(labels, mae, marker='o', label='8M Model')
plt.xlabel('Projection Dimension')
plt.ylabel('MAE')
# Log2 scale for x-axis
plt.xscale('log', base=2)
plt.xticks(labels, labels)
plt.grid()
plt.title("Projection Dimension vs MAE for 8M Model")
```

```python
# For the 35M model
labels = [64, 128, 256, 512]
mae = [0.0602, 0.0468, 0.0451, 0.0386]
plt.plot(labels, mae, marker='o', label='35M Model')
plt.xlabel('Projection Dimension')
plt.ylabel('MAE')
# Log2 scale for x-axis
plt.xscale('log', base=2)
plt.xticks(labels, labels)
plt.grid()
plt.title("Projection Dimension vs MAE for 35M Model")
```
